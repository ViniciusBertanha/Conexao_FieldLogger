```python
from tkinter import *
import xml.etree.ElementTree as ET
import requests
import pyodbc
import datetime
import subprocess, schedule, time

root = Tk()


class Funcs:

    def conexao_sql(self):
        ##### WINDOWS
        try:
            self.server = 'servername'
            self.database = 'db'
            self.username = 'sa'
            self.password = 'senha'
            self.string_conexao = 'Driver={SQL Server Native Client 11.0};Server=' + self.server + ';Database=' + \
                                  self.database + ';UID=' + self.username + ';PWD=' + self.password
            self.conexao = pyodbc.connect(self.string_conexao)
        except Exception as err:
            self.msg = f'Erro ao conectar! {err.__class__} , {err.__cause__}'
        else:
            self.msg = 'Conectado com sucesso!'
            return self.conexao.cursor()

        ##### LINUX
        # try:
        #     self.server = 'serverip'
        #     self.database = 'bd'
        #     self.username = 'sa'
        #     self.port = '59690'
        #     self.password = 'senha'
        #     self.string_conexao = 'Driver={ODBC Driver 17 for SQL Server};PORT=' + self.port + ';Server=' + self.server + ';Database=' + self.database + ';UID=' + self.username + ';PWD=' + self.password
        #     self.conexao = pyodbc.connect(self.string_conexao)
        # except Exception as err:
        #     self.msg = f'Erro ao conectar! {err.__class__} , {err.__cause__}'
        # else:
        #     self.msg = 'Conectado com sucesso!'
        #     return self.conexao.cursor()

    def desconexao(self):
        self.conexao.close()

    def atualiza_tela(self):
       root.after(1000, root.destroy) # destroy after 1 second


class App(Funcs):
    def __init__(self):
        self.root = root
        self.datetime()
        self.tela()
        self.getter()

        #self.update_tela()
        self.atualiza_tela()
        root.mainloop()

    def datetime(self):
        self.dataatual = datetime.date.today()
        self.data = self.dataatual.strftime('%d/%m/%Y')
        self.horaatual = datetime.datetime.now()
        self.hora = self.horaatual.strftime('%H:%M:%S')

    def ping(self):
        try:
            pd3 = subprocess.run(f'ping -n 1 {self.ip}', stdout=subprocess.PIPE)
            rest3 = pd3.stdout.split()
        except:
            return False
        else:
            self.stdoutd3 = rest3
            if self.stdoutd3[12] == b'TTL=128':
                return self.stdoutd3
            else:
                return False

    def tela(self):
        self.root.title('Erro')
        self.root.configure(background='#1e3743')
        self.root.geometry("530x160")
        self.frame = Frame(self.root, bd=4, bg='#dfe3ee', highlightbackground='#759fe6', highlightthickness=3)
        self.frame.place(relx=0.01, rely=0.04, relwidth=0.98, relheight=0.92)
        self.lb_frame = Label(self.frame, bg='#dfe3ee', font=('Helvetica', 12, 'bold'))
        self.lb_frame.place(relx=0.001, rely=0.001)
        self.frame.place(relx=0.01, rely=0.04, relwidth=0.98, relheight=0.92)
        self.lb_frame = Label(self.frame, text=f'{self.msg} \n {self.msg1} \n {self.msg2} \n {self.msg3} \n{self.msg4}',
                              bg='#dfe3ee', font=('Helvetica', 12, 'bold'))
        self.lb_frame.place(relx=0.001, rely=0.001)


    def getter(self):
        # extraindo fieldlogger dos filtros da esmaltação e gravando no banco
        self.msg = self.msg1 = self.msg2 = self.msg3 = self.msg4 = ''
        try:
            self.ip = '192.168.1.25'
            self.on1 = self.ping()
            print(self.on1[11], '1')
            if self.on1 == False:
                pass
            else:
                self.url0 = "http://192.168.1.25/channels.xml"
                self.header = {'Accept': 'application/xml'}
                self.r = requests.get(self.url0, headers=self.header)
                self.tree = ET.ElementTree(ET.fromstring(self.r.content))
                self.root = self.tree.getroot()
                self.lista001 = list()
                self.valor = "Value"

                for child in self.root.iter(self.valor):
                    self.n1 = child.text.strip().replace('.', ',')
                    self.lista001.append(self.n1)

                self.tupla = (
                self.data, self.hora, self.lista001[30], self.lista001[28], self.lista001[32], self.lista001[34],
                self.lista001[24], self.lista001[26], self.lista001[36], self.lista001[37], self.lista001[38],
                self.lista001[39], self.lista001[40], self.lista001[41], self.lista001[42], self.lista001[43],
                self.lista001[13], self.lista001[12], self.lista001[14], self.lista001[15], self.lista001[10],
                self.lista001[11], self.lista001[31], self.lista001[29], self.lista001[33], self.lista001[35],
                self.lista001[25], self.lista001[27], self.lista001[44], self.lista001[46], self.lista001[47],
                self.lista001[45], self.lista001[48], self.lista001[49], self.lista001[8], self.lista001[7],
                self.lista001[9], self.lista001[6], self.lista001[18], self.lista001[19], self.lista001[0],
                self.lista001[1], self.lista001[2], self.lista001[3], self.lista001[4], self.lista001[5],
                self.lista001[16], self.lista001[17], self.lista001[20], self.lista001[21], self.lista001[22],
                self.lista001[23])
                self.params1 = [self.tupla]

        except Exception as err:
            self.msgg1 = f'Erro na conexão com o FieldLogger! {err.__class__} , {err.__cause__}'

        finally:
            if self.on1 != False:
                try:
                    self.cursor1 = self.conexao_sql()
                    self.msg = 'Conexão realizada'
                except Exception as err:
                    self.msg1 = f'Erro na conexão! {err.__class__} , {err.__cause__}'
                else:
                    try:
                        self.sql1 = '''insert into CSV001 values (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?,
                                ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)'''
                        self.cursor1.executemany(self.sql1, self.params1)
                        self.conexao.commit()
                    except Exception as err:
                        self.msg1 = f'Erro ao Gravar Fluoreto! {err.__class__} , {err.__cause__}'
                    finally:
                        self.msg1 = f' Xml Fluoreto carregado com sucesso em {self.data} - {self.hora}'

                self.desconexao()
                self.msg4 = 'Desconexão realizada'

        # extraindo field logger dos chaminés na linha 5 e 6 e gravando no banco
        try:
            self.ip = '192.168.1.26'
            self.on2 = self.ping()
            print(self.on2[11], '2')
            if self.on2 == False:
                pass
            else:
                self.url1 = "http://192.168.1.26/channels.xml"
                self.header = {'Accept': 'application/xml'}
                self.r = requests.get(self.url1, headers=self.header)
                self.tree = ET.ElementTree(ET.fromstring(self.r.content))
                self.root = self.tree.getroot()

                self.lista003 = list()
                self.valor = "Value"

                for child in self.root.iter(self.valor):
                    self.n1 = child.text.strip()
                    self.lista003.append(self.n1)

                self.tupla = (
                self.data, self.hora, self.lista003[0], self.lista003[1], self.lista003[2], self.lista003[3],
                self.lista003[4], self.lista003[5])
                self.params3 = [self.tupla]

        except Exception as err:
            self.msgg2 = f'Erro na conexão com o FieldLogger! {err.__class__} , {err.__cause__}'

        finally:
            if self.on2 != False:
                try:
                    self.cursor1 = self.conexao_sql()
                    self.msg = 'Conexão realizada'
                except Exception as err:
                    self.msg1 = f'Erro na conexão! {err.__class__}, {err.__cause__}'
                else:
                    try:
                        self.sql3 = '''insert into CSV003_Chamine_Forno values (?, ?, ?, ?, ?, ?, ?, ?) '''
                        self.cursor1.executemany(self.sql3, self.params3)
                        self.conexao.commit()
                    except Exception as err:
                        self.msg2 = f'Erro ao Gravar Chaminé! {err.__class__} , {err.__cause__}'
                    finally:
                        self.msg2 = f' Xml Chaminé carregado com sucesso em {self.data} - {self.hora}'
                self.desconexao()
                self.msg4 = 'Desconexão realizada'

        # extraindo field logger da moagem e gravando no banco
        try:
            self.ip = '192.168.1.27'
            self.on3 = self.ping()
            print(self.on3[11], '3')
            if self.on3 == False:
                pass
            else:
                self.url2 = "http://192.168.1.27/channels.xml"
                self.header = {'Accept': 'application/xml'}
                self.r = requests.get(self.url2, headers=self.header)
                self.tree = ET.ElementTree(ET.fromstring(self.r.content))
                self.root = self.tree.getroot()
                self.lista002 = list()
                self.valor = "Value"

                for child in self.root.iter(self.valor):
                    self.n1 = child.text.strip().replace('.', ',').replace('-', '')
                    self.lista002.append(self.n1)

                self.tupla = (self.data, self.hora, self.lista002[2], self.lista002[3], self.lista002[4], self.lista002[5],
                              self.lista002[6], self.lista002[7], self.lista002[8], self.lista002[9], self.lista002[10])
                self.params2 = [self.tupla]

        except Exception as err:
            self.msgg3 = f'Erro na conexão com o FieldLogger! {err.__class__} , {err.__cause__}'

        finally:
            if self.on3 != False:
                try:
                    self.cursor1 = self.conexao_sql()
                    self.msg = 'Conexão realizada'
                except Exception as err:
                    self.msg1 = f'Erro na conexão! {err.__class__} , {err.__cause__}'
                else:
                    try:
                        self.sql2 = '''insert into CSV002 values (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?) '''
                        self.cursor1.executemany(self.sql2, self.params2)
                        self.conexao.commit()
                    except Exception as err:
                        self.msg3 = f'Erro ao Gravar Moagem! {err.__class__} , {err.__cause__}'
                    finally:
                        self.msg3 = f' Xml Moagem carregado com sucesso em {self.data} - {self.hora}'
                        print(self.on1[11], self.on2[11], self.on3[11])
            
                self.desconexao()
                self.msg4 = 'Desconexão realizada'


App()
# schedule.every(5).minutes.do(App)
#
# while True:
#     schedule.run_pending()
#     time.sleep(60)
```
